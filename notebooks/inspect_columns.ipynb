```python
import os
import pandas as pd

# Project paths
PROJECT_ROOT = r'C:\Users\shrey\Desktop\Projects\Explainable Price Anomaly Detector for Indian Second-hand Marketplace'
DATA_PATH = os.path.join(PROJECT_ROOT, 'data', 'cleaned_engineered.csv')

# Load dataset
df = pd.read_csv(DATA_PATH, low_memory=False)
df.columns = df.columns.str.strip().str.lower()

# Print dataset info and column names
print('Dataset shape:', df.shape)
print('\nAll columns in cleaned_engineered.csv:')
print(df.columns.tolist())
print('\nNumber of columns:', len(df.columns))
print('\nBasic info:')
print(df.info())

# Define expected columns from baseline model
expected_num_cols = ['km', 'car_age', 'km_per_year', 'max power delivered', 'alloy wheel size', 
                     'wheel base', 'no of cylinder', 'length', 'width', 'height', 'top speed', 
                     'acceleration', 'kerb weight', 'gross weight', 'front tread', 'rear tread', 
                     'turning radius', 'cargo volume', 'max torque delivered', 'max power at', 
                     'max torque at', 'bore']
expected_cat_cols = ['fuel', 'transmission', 'city', 'oem', 'model', 'variant', 'color', 
                     'engine type', 'owner_type', 'drive type', 'steering type', 
                     'front brake type', 'rear brake type', 'tyre type']

# Check for missing columns
missing_cols = [col for col in expected_num_cols + expected_cat_cols if col not in df.columns]
if missing_cols:
    print(f'\nMissing columns (expected but not found): {missing_cols}')
else:
    print('\nAll expected columns found.')

# Check for unexpected columns
extra_cols = [col for col in df.columns if col not in expected_num_cols + expected_cat_cols + ['listed_price']]
if extra_cols:
    print(f'\nExtra columns (in CSV but not expected): {extra_cols}')
else:
    print('\nNo extra columns found.')
```
